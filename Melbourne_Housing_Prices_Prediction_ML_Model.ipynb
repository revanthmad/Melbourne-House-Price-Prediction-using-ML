```python
# Drop the null values
data.dropna(axis=0, inplace = True)
data.head(10)
```

```python
# Prepare a list of features to be considered in the prediction
X = data[['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']]
X.head(10)
```

```python
# Check the statistical description of the data features
X.describe()
```

```python
# Declare a target prediction
y = data['Price']
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
# Create a Decision Tree model with a specific number for random states, just to ensure the same results for each run.
data_model = DecisionTreeRegressor(random_state=1)
```

```python
# Fit the model with data features 'X' and target prediction 'y'
data_model.fit(X, y)
```

```python
print("Consider the following 5 houses:")
test_data = X.head(10)
test_data
```

```python
# House Price Prediction Results
print("House Price Predictions for the following 5 houses are:")
results = data_model.predict(test_data)
print(results)
```

```python
# Validate the model's accuracy by comparing results with the actual prices
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y.head(10), results)
```

```python
# Alterative way to predict the House Prices with Model Selection - Train Test Split
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X, y, random_state = 0)
```

```python
tts_data_model = DecisionTreeRegressor()

tts_data_model.fit(train_X, train_y)

test_predictions = tts_data_model.predict(test_X)
print(mean_absolute_error(test_y, test_predictions))
```
